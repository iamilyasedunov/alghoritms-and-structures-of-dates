```python
import numpy as np 
T = []
        #предполагаемая длина хэш-таблицы
A = (np.sqrt(5) - 1)/2  #Кнут говорит, что это А - удачный выбор для хэширования с умножением

#инициализируем список длины m
def init_T(T, m):
    for i in range(m):
        T.append(-1)
    
#кодируем w в число по основанию 128
def to_int(w):
    k = 1
    num = 0
    for c in w[::-1]:
        num = num + ord(c)*k
        k = k*128
    return num

#хэш-функции, основанные на делении 
def k_d(k,i): #k - key
    return k%m+i

def k_m(k,i):
    return k%m

#функция добавления элемента в хэш-список с открытой адремацией
def hash_insert(k):        
    k = to_int(k)
    i = 0
    while(i < m):
        j = k_d(k, i)
        if T[j] == -1:
            T[j] = k
            return j
        else:
            i = i + 1
        print("colissia on place ", j," with key ", k, "and old key ",T[j])
    print("error overhead!")
    
#функция поиска элемента в хэш-списке с открытой адремацией
def hash_search(w):         
    k = to_int(w)
    i = 0
    j = k_d(k, i)
    while((T[j] != -1) and (i < m)):
        j = k_d(k, i)
        if T[j] == k:
            print(w,'is here in ',j,' hash = ',k)
            return j
        print("colissia on place ", j," with key ", k, "and old key ",T[j])
        i = i + 1
    print(w, ' is not here ha-ha classic')    
    return -1

#функция добавления элемента в хэш-список с цепочками
def hash_chainsearch(w):
    k = to_int(w)
    i = 0
    j = k_m(k, i)
    if (T[j] is not list)and(T[j] == k):
        print(w, ' is here')
        return j
    else:
        z = 0
        for c in T[j]:
            z = z + 1
            if c == k:
                print(w,'is here in ',j,z,' hash = ',k)
                return
                
#функция поиска элемента в хэш-списке с цепочками
def hash_chaininsert(k):
    k = to_int(k)
    i = 0
    j = k_m(k, i)
    if T[j] == -1:
        T[j] = []
        T[j].append(k)
        return j
    else:
        T[j].append(k)
        print("colissia on place ", j,i," with key ", k, "and old key ",T[j][i])
        return j,i+1

#генерация хэш-листа, в зависимости от параметра откр. адрессция или цепочками
def hashlist_insert(w, type_hashing):
    init_T(T, m)
    for c in w:
        if type_hashing is 'chain':
            hash_chaininsert(c)
        if type_hashing is 'openadress':
            hash_insert(c)
        if (type_hashing != 'openadress') and (type_hashing is not 'chain'):
            print('error type_hashing')
            return 0
```
